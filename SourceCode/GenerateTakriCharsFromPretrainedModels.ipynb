```python
import os
import cv2
from keras.utils import np_utils
from keras.models import Sequential, Model
from keras.layers import Input, Dense, Dropout, Activation, Flatten
from keras.layers import LeakyReLU
from keras.optimizers import Adam, RMSprop
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def generateImage(code,count):
    
    adam = Adam(lr=0.0002, beta_1=0.5)    
    g = Sequential()
    g.add(Dense(256, input_dim=100, activation=LeakyReLU(alpha=0.2)))
    g.add(Dense(512, activation=LeakyReLU(alpha=0.2)))
    g.add(Dense(1024, activation=LeakyReLU(alpha=0.2)))
    g.add(Dense(784, activation='sigmoid'))  # Values between 0 and 1
    g.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
    g.load_weights("D:\\Project Himachal\\Dataset\\Pretrained Models\\BZ\\"+str(code)+".h5")
    
    counter = 0
    while (counter<count):
        rtn,img = validate_generated(g)                
        if (rtn==1):
            save(img,"D:\\Project Himachal\\Dataset\\GAN Generated\\BZ\\",code,counter)
            counter = counter+1

    return rtn
```

```python
#TZ
#frequency = [1295,309,826,224,187,125]
#codes = [301,302,306,307,308,309,310]

#BZ
frequency = [569,446,56,69,228]
codes = [304,305,501,502,503]


```

```python
def validate_generated(g,n_ex=1):
    noise = np.random.normal(0, 1, size=(n_ex, 100))
    generated_images = g.predict(noise)    
    generated_images = generated_images.reshape(n_ex, 28, 28)
    black=0    
    data = np.zeros((28, 28),dtype = int)
    for row in range (28):
        for column in range (28):            
            if (generated_images[0][row][column]) >.9:
                black=black+1    
                data[row][column] = 255
            else:
                data[row][column] = 0
    
    
    if (black) <500:
        return 0,None
    else:        
        return 1,data
    
```

```python
def save(img,path,code,name):     
    #image = ~img
    print(path + "gan"+ str(code) + ".png")
    cv2.imwrite(path + "gan_"+ str(code) + "_"+ str(name) + ".png", img)
    
```
